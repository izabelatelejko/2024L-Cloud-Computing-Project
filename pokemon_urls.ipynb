```python
import requests
import string
import pandas as pd

def get_pokemon_image(pokemon_name):  
    for a in string.digits + string.ascii_lowercase[:6]:
        for b in string.digits + string.ascii_lowercase[:6]:
            url = f"https://archives.bulbagarden.net/media/upload/{a}/{a}{b}/{pokemon_name}.png"
            response = requests.get(url)
            if response.status_code == 200:
                print(url)
                return url
    print("not found")
    return "not found"

def get_pokemon_urls():
    df = pd.read_csv("pkmn-go.csv")
    df = df.reset_index()
    df["name"] = df["name"].str.replace(" ", "_")
    df["index"] = df["index"] + 1
    df.loc[28, "name"] = "Nidoran"
    df.loc[31, "name"] = "Nidoran"
    df.iloc[-3:, 0] = df.iloc[-3:, 0] + 3 
    df["pokemon_id"] = df["index"].astype(str).str.zfill(4) + df["name"]

    df["image_url"] = df["pokemon_id"].apply(get_pokemon_image)

    df = df[["index", "image_url"]]
    df.columns = ["id", "image_url"]
    df.to_csv("pokemon_urls.csv", index=False)

    return df
```

```python
df = get_pokemon_urls()
```
